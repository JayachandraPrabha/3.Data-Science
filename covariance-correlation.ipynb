```python
dataset[quan].corr()
```

```python
# correlation is the relation between two columns
# Correlation types: 
# Positive (input/independent variable is directly proprtional to output/independent variable)
    #perfect positive (r=1)
    #high degree of positive (r> +high or 0.95)
    #low degree of positive (r<+low or 0.54)
# Negative (input/independent variable is indirectly/inversly proprtional to output/independent variable)
    #perfect Negative (r=1)
    #high degree of Negative (r<-low or -0.54)
    #low degree of Negative (r>-high or -0.94)
# Zero (no pattern or relation ship between input/independent variable & output/independent variable)
# Relation between --> mba_p & salary --> 0.134084 (low Positive correlation, r<0.54)
```
