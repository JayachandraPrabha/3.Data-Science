```python
def pdfGraph(dataset,startrange,endrange):
    import seaborn as sns
    import matplotlib.pyplot as plt
    from scipy.stats import norm

    graph=sns.distplot(dataset,kde=True,kde_kws={'color':'Blue'},color='Green')
    plt.axvline(startrange,color='Red')
    plt.axvline(endrange,color='Red')

    sample=dataset

    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean:{}, Std_Dev:{}'.format(sample_mean,sample_std))

    dist=norm(sample_mean,sample_std)

    values=[]
    for value in range(startrange,endrange):
        values.append(value)

    probabilities=[]
    for value in values:
        probabilities.append(dist.pdf(value))

    prob=sum(probabilities)
    print('The area between the range:({},{}:{})'.format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
pdfGraph(dataset['ssc_p'],60,70)
```

```python
def pdff(dataset,startrange,endrange):
    import seaborn as sns
    import matplotlib.pyplot as plt
    from scipy.stats import norm

    graph=sns.distplot(dataset,kde=True,kde_kws={'color':'Black'},color="Red")
    plt.axvline(startrange,color='Green')
    plt.axvline(endrange,color='Green')

    sample=dataset

    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean:{},Std:{}'.format(sample_mean,sample_std))

    dist=norm(sample_mean,sample_std)

    values=[]
    for value in range(startrange,endrange):
        values.append(value)

    probabilities=[]    
    for value in values:
        probabilities.append(dist.pdf(value))

    prob=sum(probabilities)
    print('The area between the range:({},{} : {})'.format(startrange,endrange,sum(probabilities)))
    return prob

```
